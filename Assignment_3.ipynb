### Set up the dataframe

```python
import pandas as pd
import matplotlib.pyplot as plt
import random
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/sava-stew/Principles-of-Data-Science-Assignment3/refs/heads/main/diabetes.csv')
df
```

## **a.) Set a seed (to ensure work reproducability) and take a random sample of 25 observations and find the mean Glucose and highest Glucose values of this sample and compare these statistics with the population statistics of the same variable. You should use charts for this comparison.**

*Set seed and create random sample of 25 observations*

```python
random.seed(12)
sample = df.sample(n=25)
sample
```

*Find the mean and max glucose from the sample*

```python
meanGlucoseS = sample['Glucose'].mean()
maxGlucoseS = sample['Glucose'].max()
print(f'Mean glucose of sample: {meanGlucoseS}')
print(f'Max glucose of sample: {maxGlucoseS}')
```

*Find the mean and max glucose from the population*

```python
meanGlucoseP = df['Glucose'].mean()
maxGlucoseP = df['Glucose'].max()
print(f'Mean glucose of population: {meanGlucoseP}')
print(f'Max glucose of popoulation: {maxGlucoseP}')
```

*Compare the mean and max results from the sample and the population through the use of a bar plot*

```python
sampleGlucose = [meanGlucoseS, maxGlucoseS]
populationGlucose= [meanGlucoseP, maxGlucoseP]
compareGlucose = pd.DataFrame({'Sample': sampleGlucose, 'Population': populationGlucose}, index=['Mean Glucose', 'Max Glucose'])
compareGlucose.plot.bar(title='Glucose Comparison of Sample and Population')
```

## **b.) Find the 98th percentile of BMI of your sample and the population and compare the results using charts.**

*Find the 98th percentile of BMI from the sample*

```python
sample98 = sample['BMI'].quantile(0.98)
print(f'98th percentile of BMI in sample: {sample98}')
```

*Find the 98th percentile of BMI from the population*

```python
population98 = df['BMI'].quantile(0.98)
print(f'98th percentile of BMI in population: {population98}')
```

*Compare the 98th percentile results from the sample and the population through the use of a bar plot*

```python
sampleBMI = [sample98]
populationBMI = [population98]
compareBMI = pd.DataFrame({'Sample': sampleBMI, 'Population': populationBMI}, index=['98th Percentile of BMI'])
compareBMI.plot.bar(title='BMI Comparison of Sample and Population')
```

## **c.) Using bootstrap (replace = True), create 500 samples (of 150 observations each) from the population and find the average, standard devision, and percentile for Blood Pressure and compare this with these statistics from the population for the same variable. Again, you should create charts for this comparison. Report on your findings.**

*Create 500 samples of 150 observations each using bootstrap*

```python
bootstrap = pd.DataFrame()

for i in range(0, 500):
    add = df.sample(n = 150, replace = True)
    bootstrap = pd.concat([bootstrap, add])

bootstrap
```

*Find the mean, standard deviation and 98th percentile of blood pressure for the bootstrap*

```python
meanBP_bootstrap = bootstrap['BloodPressure'].mean()
standardBP_bootstrap = bootstrap['BloodPressure'].std()
bP98_bootstrap = bootstrap['BloodPressure'].quantile(0.98)

print(f'Mean Blood Pressure of Bootstrap: {meanBP_bootstrap}')
print(f'Standard Deviation of Blood Pressure of Bootstrap: {standardBP_bootstrap}')
print(f'98th Percentile of Blood Pressure of Bootstrap: {bP98_bootstrap}')
```

*Find the mean, standard deviation and 98th percentile of blood pressure for the population*

```python
meanBP_population = df['BloodPressure'].mean()
standardBP_population = df['BloodPressure'].std()
bP98_population = df['BloodPressure'].quantile(0.98)

print(f'Mean Blood Pressure of Population: {meanBP_population}')
print(f'Standard Deviation of Blood Pressure of Population: {standardBP_population}')
print(f'98th Percentile of Blood Pressure of Population: {bP98_population}')
```

*Compare the blood pressure results from the boostrap and the population through the use of a bar plot*

```python
bootstrapBP = [meanBP_bootstrap, standardBP_bootstrap, bP98_bootstrap]
populationBP = [meanBP_population, standardBP_population, bP98_population]
compareBMI = pd.DataFrame({'Bootstrap': bootstrapBP, 'Population': populationBP}, index=['Mean Blood Pressure', 'Standard Deviation of Blood Pressure', '98th Percentile of Blood Pressure'])
compareBMI.plot.bar(title='BP Comparison of Bootstrap and Population')
```

**Report on Findings:**
While the results of the mean, standard deviation, and  98th percentile of blood pressure were incredibly similar between the bootstrap and the population, there were some differences. The mean and standard deviation of blood pressure found from the bootstrap were less than the results of the population. The 98th percentile, however, was greater than that of the population. While the results of the bootstrap seem to be decently accurate, it does appear to have some variance on both sides of the actual results from the given population (greater and less than the population results based on the analysis occuring).
